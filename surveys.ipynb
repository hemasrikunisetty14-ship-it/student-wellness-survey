```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# 1. Count of students by Stress Level
stress_counts = df['Stress Level'].value_counts()
print("\nStress Level Distribution:")
print(stress_counts)

# 2. Bar plot: Stress Level vs Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Stress Level', order=stress_counts.index, palette='coolwarm')
plt.title('Student Stress Level Distribution')
plt.xlabel('Stress Level')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Optional: Group by Sleep Hours and Stress Level
sleep_stress = df.groupby(['Sleep Hours', 'Stress Level']).size().unstack(fill_value=0)
print("\nSleep Hours vs Stress Level:")
print(sleep_stress)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Count of students by Study Hours
study_counts = df['Study Hours'].value_counts()
print("\nStudy Hours Distribution:")
print(study_counts)

# Bar plot: Study Hours vs Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Study Hours', order=study_counts.index, palette='viridis')
plt.title('Student Study Hours Distribution')
plt.xlabel('Study Hours')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
